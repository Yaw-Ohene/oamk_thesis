```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from Excel file
file_path = 'Data/API_IC.REG.DURS_DS2_en_excel_v2_2012751.xls'  
sheet_name = 'Data' 
df = pd.read_excel(file_path, sheet_name=sheet_name, index_col=0)

# Clean the data by deleting unawanted rows and columns 
columns_to_delete = ['World Development Indicators','Unnamed: 2', 'Unnamed: 3','Unnamed: 4','Unnamed: 5', 'Unnamed: 6','Unnamed: 7','Unnamed: 8', 'Unnamed: 9','Unnamed: 10','Unnamed: 11','Unnamed: 12', 'Unnamed: 13','Unnamed: 14','Unnamed: 15', 'Unnamed: 16','Unnamed: 17','Unnamed: 18', 'Unnamed: 19','Unnamed: 20','Unnamed: 21','Unnamed: 22', 'Unnamed: 23','Unnamed: 24','Unnamed: 25', 'Unnamed: 26','Unnamed: 27','Unnamed: 28', 'Unnamed: 29','Unnamed: 30','Unnamed: 31','Unnamed: 32', 'Unnamed: 33','Unnamed: 34','Unnamed: 35', 'Unnamed: 36','Unnamed: 37','Unnamed: 38', 'Unnamed: 39','Unnamed: 40','Unnamed: 41','Unnamed: 42', 'Unnamed: 43','Unnamed: 44','Unnamed: 45', 'Unnamed: 46','Unnamed: 64','Unnamed: 65', 'Unnamed: 66','Unnamed: 67']  # Replace with the columns you want to delete
rows_to_delete = [0, 1]  
df.drop(index=df.index[rows_to_delete], inplace=True)
df.drop(columns=columns_to_delete, inplace=True)
    
# Transpose the DataFrame
df = df.transpose()
df.rename(columns={'Country Name': 'Year'}, inplace=True)

# Convert the 'Year' column to datetime format representing only the year
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

df.reset_index(drop=True, inplace=True)

selected_regions = ['Year','Ghana','Finland','World','Africa Western and Central']
new_df= df[selected_regions]
new_df

# Save the new DataFrame to an Excel file
output_file_path = 'data for business registration time for thesis.xlsx' 
new_df.to_excel(output_file_path, index=False)
print("New DataFrame has been saved to", output_file_path)

plt.figure(figsize=(10, 5))
for column in new_df.columns[1:]:  # Skip the 'Year' column for plotting
    plt.plot(new_df['Year'], new_df[column], label=column)
plt.xlabel('Year')
plt.ylabel('Number of Days')
plt.title('Time To Register A Business')
plt.legend()
plt.grid(True)
plt.show()
```
